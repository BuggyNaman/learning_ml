# List
- should be included in square brackets.
- It can have multiple data types
- Allows duplicate values

```python
my_list = [1,2,3,4,5]
print(my_list)
print(type(my_list))

multiple_data_type_list = [1, "string", 1.5, True]
print(type(multiple_data_type_list))
```

## Adding elements in a list
- keyword: append -> adds element at the last index

```python
list1 = [1,2,3,4,5]
print(list1)
list1.append(True)
print(list1)
```

## Accessing elements of a list
- method: using index


```python
list2 = [1,2,3,4]
print(list2[0])
print(list2[-1]) ## - before an index indicates start from back, -1: last element, -2:second-last and so on
print(list2[-2])
```

```python
print(len(list2))
```

# if elif else

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))
if(a<b>c):
  print(f"{b} is the largest number")
elif(b<a>c):
  print(f"{a} is the largest number")
else:
  print(f"{c} is the largest number")
```

# Functions
- keyword: def

```python
def factorial(n):
  pro = 1;
  for i in range(1,n+1):
    pro *= i
  return pro

print(factorial(5))
```
