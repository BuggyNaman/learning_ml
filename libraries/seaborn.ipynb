# Seaborn
- Data visualization library

```python
#import 
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## importing already present datasets in seaborn

```python
# datasset 'tips': it gives the data of bill vs tips
tips = sns.load_dataset('tips')
tips.head()
```

# Relative plot
- plotting to show relative tipping during lunch and dinner times

```python
sns.set_theme()
sns.relplot(data=tips, x = 'tip', y= 'total_bill', hue = 'smoker', style = 'smoker', col='time',size = 'size')
```

## Scatter plot

```python
#loading iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.scatterplot(data = iris, x='sepal_length', y = 'petal_length',hue = 'species')
```

## Count plot

```python
#loading titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# count of people in first,second and third class. Further distinguishing the people in each class based on their sex
sns.countplot(x='class',data=titanic,hue='sex')
```

```python
sns.countplot(x='survived',hue='sex',data=titanic)
```

```python
# no of people survived from each class
sns.countplot(x='class',hue='survived',data=titanic)
```

## Bar plot


```python
sns.barplot(x='sex',y='survived',hue='class',data=titanic)
```

```python
#housing dataset
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
housing
```

```python
housing_df = pd.DataFrame(housing.data, columns = housing.feature_names)
housing_df.head()
```

```python
housing_df['Price'] = housing.target
housing_df.head()
```

## Distribution plot

```python
#find the distribution of prices of the house

sns.histplot(housing_df['Price'],kde = True)
```

## Correlation matrix

```python
corr = housing_df.corr()
```

```python
fig = plt.figure(figsize = (10,5))
sns.heatmap(corr,square = True, cbar = True,cmap = 'Blues',fmt='.1f',annot = True, annot_kws={'size':8})
```
