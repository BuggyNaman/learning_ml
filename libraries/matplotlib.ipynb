# Matplotlib
- Useful for making plots

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Sine and Cosine waves

```python
x = np.linspace(0,10,100) # 100 evenly spaced values between 0,10
y = np.sin(x) # sin values for each angle in the array x
z = np.cos(x) # cos values for each angle in the array x
```

```python
print(x)
print()
print(y)
print()
print(z)
```

```python
plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('sine values')
plt.title('sine wave')
plt.show()
```

```python
plt.plot(x,z)
plt.xlabel('Angle')
plt.ylabel('cosine values')
plt.title('cosine wave')
plt.show()
```

## Parabola curve

```python
x = np.linspace(-10,10,100)
y = x**2
plt.plot(x,y)
plt.show()
```

Default is a line plot

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

## Plotting multiple graphs in one

```python
x = np.linspace(-5,5,100)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

## Bar Plot

```python
languages = ['Hindi','English','Marathi','Gujarati']
people = np.random.randint(1,250,(4))
print(people)
```

```python
fig = plt.figure(figsize=(4,4)) # creating an emplty plot
ax = fig.add_axes([0,0,1,1]) # graph starts at (0,0) and has a ht width of 1
ax.bar(languages,people,color='#134')
plt.xlabel('Languages')
plt.ylabel('No of people')
plt.show()
```

## Pie chart

```python
fig1 = plt.figure(figsize=(4,4)) # creating an emplty plot
ax = fig1.add_axes([0,0,1,1]) # graph starts at (0,0) and has a ht width of 1
ax.pie(people,labels = languages, autopct = '%1.1f%%')
# plt.xlabel('Languages')
# plt.ylabel('No of people')
plt.show()
```

## Scatter plot

```python
angle = np.linspace(5,10,50)
fig2 = plt.figure(figsize=(5,3))
ax = fig2.add_axes([0,0,1,1])
ax.scatter(angle,np.sin(angle),color='g')
ax.scatter(angle,np.cos(angle),color='b')
plt.xlabel('angle')
plt.ylabel('sin/cos values')
plt.legend(['sine', 'cos'])
plt.plot()
```

## 3D scatter plot

```python
fig3 = plt.figure(figsize = (5,5))
ax = plt.axes(projection = '3d')
x = 20*np.random.random(1000)
y = np.sin(x)
z = np.cos(x)
ax.scatter(y,z,x,c=x,cmap='Blues')
plt.show()
```
