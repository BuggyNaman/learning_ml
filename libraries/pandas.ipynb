# Manipulating the dataframe

The housing dataset has a feature called target but it is not included in dataframe

## adding column in dataframe
- Note: to this no of data points must be same.
- The og housing dataset had 20640 data points. Hence, the target column must also have 20640 data points


```python
housing_df.shape
```

```python
# creating a new column called price

housing_df['Price'] = housing.target
housing_df.head()
```

## Removing rows and columns

```python
housing_df.head()
```

```python
#removing row (give index i.e. which row number and axis = 0 indicates row removal)
housing_df.drop(index=1,axis=0,inplace = True)
housing_df.head()
```

```python
# removing column. Give column name in and axis = 1 indicates column removal
housing_df.drop(columns='Population',axis=1,inplace = True)
housing_df.head()
```

## Loacting or accessing a particular row using index value

```python
housing_df.iloc[2] # gives each column value for row indexed 2
```

```python
print(housing_df.iloc[:,0]) # first column
print()
print(housing_df.iloc[:,1]) # second column
print()
print(housing_df.iloc[:,2]) # third column
print()
print(housing_df.iloc[:,-1]) # last column
```
